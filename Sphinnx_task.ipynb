```python
import moviepy.editor as mp
#  function to extract aud from vid
def extract_audio(video_path, audio_output_path):
    video = mp.VideoFileClip(video_path)
    audio = video.audio
    audio.write_audiofile(audio_output_path)



video_path = r"/content/sample_data/Grit.mp4"


audio_output_path = "/content/sample_data/audio.wav"
translated_audio_path = "/content/hello.mp3"

#extracting audio from vidd
extract_audio(video_path, audio_output_path)

```


***Here I have written code that is translating audio to text and then text to audio in english to hindi which is playing the audio in hindi from english.***
*  Loading the audio file
*  Recognizing the audio

*  Translating the text
*  Converting translated text to audio

*   Finally audio play








```python
import googletrans
import speech_recognition
import gtts
from IPython.display import Audio
import os

audio_output_path = "/content/sample_data/audio.wav"

recognizer = speech_recognition.Recognizer()


with speech_recognition.AudioFile(audio_output_path) as source:
    audio_data = recognizer.record(source)


text = recognizer.recognize_google(audio_data, language="en")
print("Recognized Text:", text)

#trnslating to text
translator = googletrans.Translator()
translation = translator.translate(text, dest="hi")
print("Translated Text:", translation.text)

# txt to aud
converted_audio = gtts.gTTS(translation.text, lang="hi")
converted_audio_path = "hello.mp3"
converted_audio.save(converted_audio_path)

# play aud
Audio(converted_audio_path, autoplay=True)
```

***Here i have written code that is integrating my translated audio to my video.***

*   Loading the MP4 video
*   Loading the converted audio
*   Set the audio of the video to the converted audio
*   Write the new video with the integrated audio




```python
from moviepy.editor import VideoFileClip, AudioFileClip


video_clip = VideoFileClip(video_path )


audio_clip = AudioFileClip(translated_audio_path)


video_clip = video_clip.set_audio(audio_clip)


video_clip.write_videofile("output_video.mp4", codec="libx264", audio_codec="aac")
```
